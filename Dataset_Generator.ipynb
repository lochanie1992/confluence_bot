```python
import pandas as pd
import random
from faker import Faker
import numpy as np

# Initialize Faker to generate fake data
fake = Faker()

# Create sample data for the fields
titles = [
    "How to Set Up an ETL Pipeline in Python",
    "Best Practices for Data Modeling in Data Warehouses",
    "What is a Data Lake and How to Use It?",
    "SQL vs NoSQL: Which Database to Use?",
    "Data Analytics with Python: A Beginner's Guide",
    "Introduction to Power BI for Data Analysis",
    "Data Warehousing and Cloud Computing Integration",
    "Building Real-Time Data Pipelines with Kafka",
    "Data Security Best Practices for Cloud Databases",
    "Machine Learning with Data Lakes"
]

# Randomly generate 1000 records
data = {
    "title": [],
    "content": [],
    "resources": [],
    "video_links": [],
    "web_links": [],
    "author": [],
    "people_tagged": [],
    "modified_date": []
}

for _ in range(1000):
    # Randomly pick a title and create content based on the title
    title = random.choice(titles)
    content = f"Content for {title}. This is a placeholder for the detailed explanation about {title}. It covers the concepts, techniques, and best practices."
    
    # Randomly generate resources and links
    resources = f"https://www.example.com/{fake.word()}-guide"
    video_links = f"https://www.youtube.com/watch?v={fake.bothify(text='??#####')}"
    web_links = f"https://www.example.com/{fake.word()}"
    
    # Random authors and people tagged
    author = fake.name()
    people_tagged = f"{fake.name()}; {fake.name()}"
    
    # Random modification date
    modified_date = fake.date_this_year()
    
    # Append to the dataset
    data["title"].append(title)
    data["content"].append(content)
    data["resources"].append(resources)
    data["video_links"].append(video_links)
    data["web_links"].append(web_links)
    data["author"].append(author)
    data["people_tagged"].append(people_tagged)
    data["modified_date"].append(modified_date)

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("confluence_pages.csv", index=False)

print("Dataset generated and saved as confluence_pages.csv.")
```
